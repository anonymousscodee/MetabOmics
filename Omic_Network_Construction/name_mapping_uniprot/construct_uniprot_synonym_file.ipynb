```python
from collections import defaultdict

# Load HGNC file
hgnc_df = pd.read_csv("hgnc_complete_set.txt", sep="\t", dtype=str).fillna("")

# Final mapping: name/synonym → list of UniProt IDs
hgnc_to_uniprot = defaultdict(list)

for _, row in hgnc_df.iterrows():
    uniprot_ids = row['uniprot_ids'].split('|') if row['uniprot_ids'] else []
    if not uniprot_ids:
        continue

    # 1. HGNC ID
    hgnc_id = row['hgnc_id']
    if hgnc_id:
        hgnc_to_uniprot[hgnc_id].extend(uniprot_ids)

    # 2. Approved Symbol
    approved_symbol = row['symbol']
    if approved_symbol:
        hgnc_to_uniprot[approved_symbol].extend(uniprot_ids)

    # 3. Approved Name
    approved_name = row['name']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 4. Approved entrez_id
    approved_name = row['entrez_id']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 5. Approved ensembl_gene_id
    approved_name = row['ensembl_gene_id']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 6. Approved vega_id
    approved_name = row['vega_id']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 7. Approved ucsc_id
    approved_name = row['ucsc_id']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 8. Approved ena
    approved_name = row['ena']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 9. Approved ccds_id
    approved_name = row['ccds_id']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 10. Approved uniprot_ids
    approved_name = row['uniprot_ids']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

     # 11. Approved mgd_id
    approved_name = row['mgd_id']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 12. Approved omim_id
    approved_name = row['omim_id']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 13. Approved mirbase
    approved_name = row['mirbase']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)
    
    # 14. Approved alias_symbol
    approved_name = row['alias_symbol']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)
    
    # 15. Approved alias_name
    approved_name = row['alias_name']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

     # 16. Approved prev_symbol
    approved_name = row['prev_symbol']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 17. Approved prev_name
    approved_name = row['prev_name']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)

    # 18. Approved refseq_accession
    approved_name = row['refseq_accession']
    if approved_name:
        hgnc_to_uniprot[approved_name].extend(uniprot_ids)
# Optional: Remove duplicates from values
for key in hgnc_to_uniprot:
    hgnc_to_uniprot[key] = sorted(set(hgnc_to_uniprot[key]))
```

```python
pivot_df['UniProtKB-ID'] = pivot_df['UniProtKB-ID'].str.replace('_HUMAN$', '', regex=True)
pivot_df
```

```python
'''filtered_pivot = pivot_df[['UniProt_ID','Gene_Name', 'CCDS', 'Ensembl','GeneID', 'KEGG', 'HGNC', 'VEuPathDB',  
                           'Gene_Synonym', 'ChiTaRS', 'RefSeq', 'UniProtKB-ID', 'GeneTree', 'DrugBank', 'ChEMBL', 
                           'BioMuta', 'OMA']]'''
filtered_pivot = pivot_df[['UniProt_ID','Gene_Name','Ensembl','KEGG', 'HGNC','UniProtKB-ID', 'STRING','GeneID', 
                           'Gene_Synonym', 'Gene_ORFName', 'GeneCards']]
filtered_pivot
```

```python
value_to_uniprot = {}
for column in filtered_pivot.columns:
    if column == 'UniProt_ID':
        continue  # skip the 'UniProt_ID' column itself

    # Only keep rows where this column is not NaN
    non_empty = filtered_pivot.loc[filtered_pivot[column].notna(), ['UniProt_ID', column]]

    for _, row in non_empty.iterrows():
        uniprot_id = row['UniProt_ID']
        ext_ids = str(row[column]).split(';')

        # Add mapping: ext_id → uniprot_id
        for ext_id in ext_ids:
            ext_id = ext_id.strip()
            if not ext_id:
                continue
            if ext_id in value_to_uniprot:
                value_to_uniprot[ext_id].add(uniprot_id)
            else:
                value_to_uniprot[ext_id] = {uniprot_id}

        # Add self-mapping: uniprot_id → uniprot_id
        if uniprot_id in value_to_uniprot:
            value_to_uniprot[uniprot_id].add(uniprot_id)
        else:
            value_to_uniprot[uniprot_id] = {uniprot_id}         
#value_to_uniprot                
```

```python
count = 0
for key, val in value_to_uniprot.items() :
    if key in ['CYTB', 'ND5', 'ATP6', 'ND2'] :
        continue
    if count < 6 :
        count += 1
        print(f"{key} : {val}")
    else : break
```

```python
# Merge KEGG-HGNC-UniProt Dictionaries
# hgnc_to_uniprot + kegg_to_uniprot + value_to_uniprot
from collections import defaultdict
import json

# Initialize merged dictionary with default value as empty set
merged_dict = defaultdict(set)

# List of dictionaries to merge
#all_dicts = [hgnc_to_uniprot, kegg_to_uniprot, value_to_uniprot]
all_dicts = [hgnc_to_uniprot, value_to_uniprot]
# Merge them
for d in all_dicts:
    for k, v in d.items():
        merged_dict[k].update(v)
```

```python
len(merged_dict)
```

```python
# Convert sets to lists since JSON does not support sets
json_ready_dict = {k: list(v) for k, v in merged_dict.items()}

# Save to JSON file
with open("uniprot_synonym_mapping.json", "w", encoding="utf-8") as f:
    json.dump(json_ready_dict, f, indent=2)
```
