```python
cleaned_data_uniprot = {k.split("_", 1)[0]: v for k, v in manual_id_mapping['u_uniprot_acc'].items()}
len(cleaned_data_uniprot)
```

```python
not_found = set()
found = {}
for g in modelREcon3d.genes :
    g_id = g.id.split("_", 1)[0]
    if g_id in cleaned_data_uniprot :
        found[g.id] = cleaned_data_uniprot[g_id]
    else :
        not_found.add(g.id)
print("Number of not founded genes : ", len(not_found))
```
